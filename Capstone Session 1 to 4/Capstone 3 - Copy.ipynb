```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the CSV file into a DataFrame
df = pd.read_csv('NSMES1988updated.csv')

```

```python
# Display basic info about the DataFrame
print("DataFrame loaded successfully")

# prints the size of a DataFrame (df) using an f-string.
# df.shape returns a tuple → (number of rows, number of columns)
# df.shape[1] specifically gets the number of columns, df.shape[0] is # of row as index in Python functions def.shape

print(f"Shape: {df.shape} rows x {df.shape[1]} columns")
print("\nFirst 5 rows:")
print(df.head())

```

```python
# Identify different types of data and report it
print("\n" + "="*50)
print("DATA TYPES REPORT")
print("="*50)
print(df.dtypes)

# Count of each data type
print("\nData type counts:")
print(df.dtypes.value_counts())

# Count row of same values - for duplicates???? 
print("\nData type counts DUPLICATE ROWS:")
print(df.value_counts())


print('WHAT THE HECK IS GOING ON: this is a test on 05/01/2026 3:54pm')
```

```python
# Identify categorical types in the data
print("\n" + "="*50)
print("CATEGORICAL DATA TYPES")
print("="*50)
```

```python
# Method 1: Columns with 'object' dtype (typically categorical/text)

object_columns = df.select_dtypes(include=['object']).columns

print(f"\nObject type columns (likely categorical): {len(object_columns)}")
for col in object_columns:
    unique_vals = df[col].nunique()
    print(f"  - {col}: {unique_vals} unique values")

# Method 2: Columns with few unique values (could be categorical)
print("\nPotential categorical columns (few unique values):")
for col in df.columns:
    if df[col].dtype != 'object' and df[col].nunique() < 20:
        print(f"  - {col}: {df[col].nunique()} unique values (dtype: {df[col].dtype})")

```

```python
# Basic pivot table (if 'age' and 'income' columns exist)
if 'age' in df.columns and 'income' in df.columns:
    print("\n1. Pivot Table: Average income by age group")
    # Create age groups
    df['age_group'] = pd.cut(df['age'], bins=[0, 18, 30, 45, 60, 100], 
                              labels=['0-18', '19-30', '31-45', '46-60', '60+'])
    pivot1 = pd.pivot_table(df, values='income', index='age_group', aggfunc='mean')
    print(pivot1)
```

```python
# Cross-tabulation (if categorical columns exist)
if len(object_columns) > 0:
    cat_col = object_columns[0]
    print(f"\n2. Cross-tabulation for {cat_col} (first 5 values)")
    crosstab = pd.crosstab(df[cat_col].fillna('Missing'), df['age_group'] if 'age_group' in df.columns else df.index)
    print(crosstab.head())


# Summary pivot table for numeric columns
print("\n4. Summary Pivot: Mean of all numeric columns by age_group")
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
if 'age_group' in df.columns and numeric_cols:
    pivot_summary = pd.pivot_table(df, values=numeric_cols, index='age_group', aggfunc='mean')
    print(pivot_summary)
```

```python
# Final report summary
print("\n" + "="*50)
print("FINAL REPORT SUMMARY")
print("="*50)
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"Numeric columns: {len(numeric_cols)}")
print(f"Object (categorical-like) columns: {len(object_columns)}")
print(f"Age groups created: {df['age_group'].nunique() if 'age_group' in df.columns else 0}")
```

```python
# Detailed data pivoting focusing on Health and Region
print("\n" + "="*70)
print("4. DATA PIVOTING ANALYSIS (Focus on Health & Region)")
print("="*70)

# Pivot 1: Average income by Health and Region (if income exists)
if 'income' in df.columns and 'health' in df.columns and 'region' in df.columns:
    print("\nA. Average Income by Health Status and Region")
    pivot_health_region = pd.pivot_table(df, values='income', 
                                         index='health', 
                                         columns='region', 
                                         aggfunc='mean', 
                                         fill_value=0)
    print(pivot_health_region)
  
```

```python
# Pivot 2: Health distribution across Regions
if 'health' in df.columns and 'region' in df.columns:
    print("\nC. Health Status Distribution by Region (percentages)")
    health_region_cross = pd.crosstab(df['health'], df['region'], normalize='columns') * 100
    print(health_region_cross.round(1))
```

```python
# Pivot 3: If age groups exist, analyze Health by age and Region
if 'age_group' in df.columns and 'health' in df.columns and 'region' in df.columns:
    print("\nD. Average Income by Health, Age Group, and Region")
    pivot_multi = pd.pivot_table(df, values='income', 
                                 index=['health', 'age_group'],
                                 columns='region',
                                 aggfunc='mean', 
                                 fill_value=0)
    print(pivot_multi.head(10))
```
